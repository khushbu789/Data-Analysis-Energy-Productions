```python
import pandas as pd
```

```python
df=pd.read_csv('Energy_production.csv')
print(len(df))
```

```python
df.drop_duplicates()
df
```

```python
df.replace("",0)
df
```

```python
df.dropna()
df
```

```python
import numpy as np

# print(type(df['Nuclear Generation Actual (in MU)'][3]))
df=df.replace(np.nan,0)


df['Year']=df['Date'][6:10]
df
# df=df.dropna()
# df
```

```python
import matplotlib.pyplot as plt
```

```python
for i in range(len(df['Date'])):
    df['Year'][i]=df['Date'][i][6:10]
    
df
```

```python
df=df.rename(columns={
    'Thermal Generation Actual (in MU)':'TGA',
    'Thermal Generation Estimated (in MU)':'TGE',
    'Nuclear Generation Actual (in MU)':'NGA',
    'Nuclear Generation Estimated (in MU)':'NGE',
    'Hydro Generation Actual (in MU)':'HGA',
    'Hydro Generation Estimated (in MU)':'HGE',
})
```

```python
# print(df['TGA'][0])
# print(df[(df["Region"]=="Western") & (df["Year"]=="2017")]['TGE'])

plt.plot(list(df[(df["Region"]=="Western") & (df["Year"]=="2017")]["Date"][0:5]),list(df[(df["Region"]=="Western") & (df["Year"]=="2017")]["TGA"][0:5]),label='TGA')
plt.plot(list(df[(df["Region"]=="Western") & (df["Year"]=="2017")]["Date"][0:5]),list(df[(df["Region"]=="Western") & (df["Year"]=="2017")]["TGE"][0:5]),label="TGE")
plt.legend
plt.show()
```

```python
plt.plot(list(df[(df["Region"]=="Eastern") & (df["Year"]=="2019")]["Date"][0:5]),list(df[(df["Region"]=="Eastern") & (df["Year"]=="2019")]["HGA"][0:5]),label='HGA')
plt.plot(list(df[(df["Region"]=="Eastern") & (df["Year"]=="2019")]["Date"][0:5]),list(df[(df["Region"]=="Eastern") & (df["Year"]=="2019")]["HGE"][0:5]),label="HGE")
plt.legend()
plt.show()
```
